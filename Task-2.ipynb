```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


titanic_data = pd.read_csv("tested.csv")
titanic_data.head(10)

```

```python
# Check for missing values
missing_values = titanic_data.isnull().sum()
display(missing_values)



```

```python

titanic_data = titanic_data.drop(["Fare"], axis=1)

display(titanic_data)
```

```python
titanic_data=titanic_data.dropna()
display(titanic_data)

```

```python
titanic_data.shape
```

```python
data = pd.DataFrame(titanic_data.dtypes)
data
```

```python
sns.lineplot(data=titanic_data.drop(['Age'], axis=1))
plt.show()
```

```python
sns.barplot(x='Age',y='Sex', data=titanic_data)

plt.show()
```

```python
sns.countplot(x='Survived',data=titanic_data)
plt.title('Survival Count')
plt.show()
```

```python

sns.countplot(x='Survived',hue='Pclass', data=titanic_data)
plt.title('Survival by class')
plt.show()
```

```python
sns.histplot(x='Age', data=titanic_data, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(x='Fare', data=titanic_data, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
sns.boxplot(x='Survived',y='Age',data=titanic_data)
plt.title('Survival by age')
plt.show()
```

```python


sns.countplot(x='Survived', hue='SibSp', data=titanic_data)
plt.title('Survival by siblings/spouses aboard')
plt.show()

```

```python
survival_rate=titanic_data.groupby(['Pclass','Sex'])['Survived'].mean().reset_index()
survival_rate_age=titanic_data.groupby('AgeGroup')['Survived'].mean().reset_index()
sns.barplot(x='AgeGroup', y='Survived', data=survival_rate_age)
plt.title('Survival_rate_age')
plt.show()
```

```python
sns.countplot(x='Survived', hue='FamilySize', data=titanic_data)
plt.title('Survival by Family Size')
plt.show()
```
